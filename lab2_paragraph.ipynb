# unique bigrams, trigrams and quadgrams

bigrams, trigrams, and quadgrams are types of n-grams, which are contiguous sequences of n items (usually words) from a given text or corpus. This help in capturing the contextual information and patterns within a text.

```python
# Tokenize the paragraph into words
tokens = nltk.word_tokenize(para)

# Find bigrams, trigrams, and quadgrams
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))
quadgrams = list(ngrams(tokens, 4))

# Count unique bigrams, trigrams, and quadgrams
unique_bigrams = Counter(bigrams)
unique_trigrams = Counter(trigrams)
unique_quadgrams = Counter(quadgrams)

print("List of unique bigrams:", bigrams)
print("No. of unique bigrams:", len(unique_bigrams))

print("List of unique trigrams:", trigrams)
print("No. of unique trigrams:", len(unique_trigrams))

print("List of unique quadgrams:", quadgrams)
print("No. of unique quadgrams:", len(unique_quadgrams))
```

#dates

```python
# Function to convert dates to DD-MM-YYYY format
def convert_to_dd_mm_yyyy(date_str):
    try:
        date_obj = datetime.strptime(date_str, '%B-%d-%Y')
        return date_obj.strftime('%d-%m-%Y')
    except ValueError:
        return date_str

# Regular expression to find dates in the paragraph
date_pattern = r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December)-\d{1,2}-\d{4}\b'

# Replace dates in the paragraph with the formatted dates
formatted_para = re.sub(date_pattern, lambda match: convert_to_dd_mm_yyyy(match.group()), para)

print("Original Paragraph:")
print(para)
print("\nParagraph with Dates Converted to DD-MM-YYYY Format:")
print(formatted_para)
```

# Plotting year

```python
# Regular expression pattern to find years in the format 'YYYY'
pattern = r'\b\d{4}\b'

# Find all occurrences of the pattern in the paragraph
years = re.findall(pattern, para)

# Count the occurrences of each year
year_counts = {}
for year in years:
    year_counts[year] = year_counts.get(year, 0) + 1

# Plotting the distribution
plt.bar(year_counts.keys(), year_counts.values())
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of Years')
plt.show()
```

#Phone Number

```python
def is_valid_phone_number(phone_number):
    pattern = r'\b(?:\+?1[ -]?)?\(?\d{3}\)?[ -]?\d{3}[ -]?\d{4}\b'
    return re.match(pattern, phone_number) is not None

def remove_invalid_phone_numbers(text):
    # Regular expression pattern to find phone numbers
    phone_number_pattern = r'\b(?:\+?1[ -]?)?\(?\d{3}\)?[ -]?\d{3}[ -]?\d{4}\b'

    # Find all potential phone numbers in the text
    potential_phone_numbers = re.findall(phone_number_pattern, text)

    # Validate each phone number and remove invalid ones from the text
    for phone_number in potential_phone_numbers:
        if not is_valid_phone_number(phone_number):
            # Remove only if the phone number doesn't match the valid pattern
            text = text.replace(phone_number, '')

    return text

# Test the function
para = """
In the realm of data science, mastering Python is essential; its not just about understanding the syntax, but also comprehending its vast ecosystem. With libraries like NumPy, pandas, and scikit-learn, data manipulation becomes intuitive, enabling efficient analysis and modeling. Dont overlook the importance of data visualization; tools like Matplotlib and Seaborn transform raw numbers into insightful plots, aiding in the communication of findings. Remember, practice is key; consistent coding sessions hone your skills and solidify concepts. For any inquiries, feel free to reach out via email at paragraph@email.com or call me at 9499719173 or 12345. And yes, my date of birth is March-27-2000. And yes, my date sister  birthday is on 31-April-2005.  my joining date of job is May-27-2020.
"""
cleaned_text = remove_invalid_phone_numbers(para)
print("Cleaned Text:")
print(cleaned_text)
```
